```python
inventory = {
    "apple": 10,
    "banana": 5,
    "orange": 8,
    "grape": 15,
    "mango": 12,
    "cranberry": 7
}
```

```python
def display_inventory(inventory):
    print("Current Inventory:")
    for fruit, quantity in inventory.items():
        print(f"{fruit}: {quantity} units")
```

```python
display_inventory(inventory)
```
